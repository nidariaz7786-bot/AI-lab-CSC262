7. Display all attributes method

---



```python
class Student:
    def __init__(self, sid, name, sclass):
        self.student_id = sid
        self.student_name = name
        self.student_class = sclass

    def display(self):
        for key, value in self.__dict__.items():
            print(key, ":", value)

s = Student(1, "Ali", "10th")
s.display()
```

9. Memory allocation analysis

---



```python
import sys

class Student:
    def __init__(self, name):
        self.name = name

s = Student("Ali")

print("Size of object:", sys.getsizeof(s))
print("Size of dict:", sys.getsizeof(s.__dict__))
```

10. Reduce memory usage (__slots__)

---



```python
class Student:
    __slots__ = ['name']

    def __init__(self, name):
        self.name = name

s = Student("Nida")

import sys
print("Reduced size:", sys.getsizeof(s))
```

# **Lab Task 2: Object-Oriented Design and Area Comparison**

1. Classes for Trapezoid and Parallelogram

---



```python
class Trapezoid:
    def __init__(self, a, b, height):
        self.a = a
        self.b = b
        self.height = height

    def area(self):
        return 0.5 * (self.a + self.b) * self.height


class Parallelogram:
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return self.base * self.height



t = Trapezoid(4, 6, 5)
p = Parallelogram(5, 4)

print("Trapezoid Area:", t.area())
print("Parallelogram Area:", p.area())
```

2. Comparison Mechanism for Areas

---



```python
t = Trapezoid(4, 6, 5)
p = Parallelogram(5, 4)

if t.area() > p.area():
    print("Trapezoid has larger area")
elif t.area() < p.area():
    print("Parallelogram has larger area")
else:
    print("Both have equal area")
```

3. Store Areas in Dictionary

---



```python
areas = {}

t = Trapezoid(4, 6, 5)
p = Parallelogram(5, 4)

areas["Trapezoid"] = t.area()
areas["Parallelogram"] = p.area()

print(areas)
```

4. Iterate Over Multiple Dimensions

---



```python
trapezoids = [(4, 6, 5), (3, 5, 4), (6, 8, 7)]
parallelograms = [(5, 4), (6, 3), (7, 6)]

for a, b, h in trapezoids:
    t = Trapezoid(a, b, h)
    print("Trapezoid Area:", t.area())

for b, h in parallelograms:
    p = Parallelogram(b, h)
    print("Parallelogram Area:", p.area())
```

5. Find Shape with Maximum Area Using Dictionary

---



```python
areas = {}

trapezoids = [(4, 6, 5), (3, 5, 4)]
parallelograms = [(5, 4), (7, 6)]


for i, (a, b, h) in enumerate(trapezoids, 1):
    t = Trapezoid(a, b, h)
    areas[f"Trapezoid_{i}"] = t.area()

for i, (b, h) in enumerate(parallelograms, 1):
    p = Parallelogram(b, h)
    areas[f"Parallelogram_{i}"] = p.area()


max_shape = max(areas, key=areas.get)

print("All Areas:", areas)
print("Maximum Area Shape:", max_shape)
print("Area:", areas[max_shape])
```
