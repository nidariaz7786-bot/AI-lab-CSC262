```python
graph = {
    '5': ['3', '7'],
    '3': ['2', '4'],
    '7': ['8'],
    '2': [],
    '4': ['8'],
    '8': []
}

visited = set()

def dfs(visited, graph, node):
    if node not in visited:
        print(node)
        visited.add(node)
        for neighbour in graph[node]:
            dfs(visited, graph, neighbour)

print("Following is the Depth-First Search")
dfs(visited, graph, '5')
```

```python
from collections import defaultdict

class Graph:

    def __init__(self):
        self.graph = defaultdict(list)

    def addEdge(self, u, v):
        self.graph[u].append(v)

    def DFSUtil(self, v, visited):
        visited[v] = True
        print(v, end=" ")

        for i in self.graph[v]:
            if not visited[i]:
                self.DFSUtil(i, visited)


    def DFS(self):
        V = max(self.graph) + 1

        visited = [False] * V

        for i in range(V):
            if not visited[i]:
                self.DFSUtil(i, visited)


g = Graph()
g.addEdge(0, 1)
g.addEdge(0, 2)
g.addEdge(1, 2)
g.addEdge(2, 0)
g.addEdge(2, 3)
g.addEdge(3, 3)

print("Following is Depth First Traversal")
g.DFS()
```

```python
from collections import deque

graph = {
    "Arad": ["Sibiu", "Timisoara", "Zerind"],
    "Sibiu": ["Arad", "Fagaras", "Rimnicu Vilcea", "Oradea"],
    "Timisoara": ["Arad", "Lugoj"],
    "Zerind": ["Arad", "Oradea"],
    "Oradea": ["Zerind", "Sibiu"],
    "Fagaras": ["Sibiu", "Bucharest"],
    "Rimnicu Vilcea": ["Sibiu", "Pitesti", "Craiova"],
    "Lugoj": ["Timisoara", "Mehadia"],
    "Mehadia": ["Lugoj", "Drobeta"],
    "Drobeta": ["Mehadia", "Craiova"],
    "Craiova": ["Drobeta", "Rimnicu Vilcea", "Pitesti"],
    "Pitesti": ["Rimnicu Vilcea", "Craiova", "Bucharest"],
    "Bucharest": ["Fagaras", "Pitesti", "Giurgiu", "Urziceni"],
    "Giurgiu": ["Bucharest"],
    "Urziceni": ["Bucharest", "Hirsova", "Vaslui"],
    "Hirsova": ["Urziceni", "Eforie"],
    "Eforie": ["Hirsova"],
    "Vaslui": ["Urziceni", "Iasi"],
    "Iasi": ["Vaslui", "Neamt"],
    "Neamt": ["Iasi"]
}

def dfs(start, goal):
    stack = [(start, [start])]
    visited = set()
    nodes_visited = 0

    while stack:
        (vertex, path) = stack.pop()
        if vertex not in visited:
            visited.add(vertex)
            nodes_visited += 1
            if vertex == goal:
                return path, nodes_visited
            for neighbor in reversed(graph.get(vertex, [])):
                stack.append((neighbor, path + [neighbor]))
    return None, nodes_visited

def bfs(start, goal):
    queue = deque([(start, [start])])
    visited = set()
    nodes_visited = 0

    while queue:
        (vertex, path) = queue.popleft()
        if vertex not in visited:
            visited.add(vertex)
            nodes_visited += 1
            if vertex == goal:
                return path, nodes_visited
            for neighbor in graph.get(vertex, []):
                queue.append((neighbor, path + [neighbor]))
    return None, nodes_visited

print("Case 1: BFS fewer nodes than DFS (Arad -> Sibiu)")
path_bfs, visited_bfs = bfs("Arad", "Sibiu")
path_dfs, visited_dfs = dfs("Arad", "Sibiu")
print("BFS Path:", path_bfs, "Nodes visited:", visited_bfs)
print("DFS Path:", path_dfs, "Nodes visited:", visited_dfs)

print("\nCase 2: BFS more nodes than DFS (Arad -> Bucharest)")
path_bfs, visited_bfs = bfs("Arad", "Bucharest")
path_dfs, visited_dfs = dfs("Arad", "Bucharest")
print("BFS Path:", path_bfs, "Nodes visited:", visited_bfs)
print("DFS Path:", path_dfs, "Nodes visited:", visited_dfs)
```
