# **Activity 01:**

```python
visited = []
queue = []

def bfs(visited, graph, node):
    visited.append(node)
    queue.append(node)

    while queue:
        m = queue.pop(0)
        print(m, end=" ")

        for neighbour in graph[m]:
            if neighbour not in visited:
                visited.append(neighbour)
                queue.append(neighbour)

# Driver Code
graph = {
    '5': ['3', '7'],
    '3': ['2', '4'],
    '7': ['8'],
    '2': [],
    '4': ['8'],
    '8': []
}

print("Following is the Breadth-First Search")
bfs(visited, graph, '5')
```

# **Activity 02:**

```python
# Function to print the parent of each node (DFS)
def printParents(node, adj, parent):
    if parent == 0:
        print(node, "-> Root")
    else:
        print(node, "->", parent)

    for cur in adj[node]:
        if cur != parent:
            printParents(cur, adj, node)


# Function to print children of each node (BFS)
def printChildren(Root, adj):
    q = []
    q.append(Root)

    vis = [0] * len(adj)

    while len(q) > 0:
        node = q[0]
        q.pop(0)

        vis[node] = 1
        print(node, "-> ", end=" ")

        for cur in adj[node]:
            if vis[cur] == 0:
                print(cur, end=" ")
                q.append(cur)

        print()


# Function to print leaf nodes
def printLeafNodes(Root, adj):
    for i in range(len(adj)):
        if len(adj[i]) == 1 and i != Root:
            print(i, end=" ")
    print()


# Function to print degrees
def printDegrees(Root, adj):
    for i in range(1, len(adj)):
        print(i, ":", end=" ")

        if i == Root:
            print(len(adj[i]))
        else:
            print(len(adj[i]) - 1)


# Driver Code
N = 7
Root = 1

adj = []
for i in range(N + 1):
    adj.append([])

# Creating tree
adj[1].append(2)
adj[2].append(1)

adj[1].append(3)
adj[3].append(1)

adj[1].append(4)
adj[4].append(1)

adj[2].append(5)
adj[5].append(2)

adj[2].append(6)
adj[6].append(2)

adj[4].append(7)
adj[7].append(4)

print("Parents:")
printParents(Root, adj, 0)

print("\nChildren:")
printChildren(Root, adj)

print("\nLeaf Nodes:")
printLeafNodes(Root, adj)

print("\nDegrees:")
printDegrees(Root, adj)
```

# **Activity 04:**

```python
from collections import deque

graph = {
    'A': [('B', 6), ('C', 9), ('E', 1)],
    'B': [('A', 6), ('D', 3), ('E', 4)],
    'C': [('A', 9), ('F', 2), ('G', 3)],
    'D': [('B', 3), ('E', 5), ('F', 7)],
    'E': [('A', 1), ('B', 4), ('D', 5), ('F', 6)],
    'F': [('C', 2), ('E', 6), ('D', 7)],
    'G': [('C', 3)]
}

def bfs_with_cost(graph, start, goal):
    queue = deque()

    queue.append((start, [start], 0))

    visited = set()

    while queue:
        node, path, cost = queue.popleft()

        if node == goal:
            print("Path:", " -> ".join(path))
            print("Total Cost:", cost)
            return

        if node not in visited:
            visited.add(node)

            # Explore neighbors
            for neighbour, weight in graph[node]:
                if neighbour not in visited:
                    queue.append((neighbour, path + [neighbour], cost + weight))

    print("No path found")

start = 'A'
goal = 'G'

print("BFS from", start, "to", goal)
bfs_with_cost(graph, start, goal)
```

```python
from collections import deque

# Graph (adjacency list)
graph = {
    'Arad': ['Zerind', 'Sibiu', 'Timisoara'],
    'Zerind': ['Arad', 'Oradea'],
    'Oradea': ['Zerind', 'Sibiu'],
    'Timisoara': ['Arad', 'Lugoj'],
    'Lugoj': ['Timisoara', 'Mehadia'],
    'Mehadia': ['Lugoj', 'Drobeta'],
    'Drobeta': ['Mehadia', 'Craiova'],
    'Craiova': ['Drobeta', 'Rimnicu Vilcea', 'Pitesti'],
    'Sibiu': ['Arad', 'Oradea', 'Fagaras', 'Rimnicu Vilcea'],
    'Fagaras': ['Sibiu', 'Bucharest'],
    'Rimnicu Vilcea': ['Sibiu', 'Craiova', 'Pitesti'],
    'Pitesti': ['Rimnicu Vilcea', 'Craiova', 'Bucharest'],
    'Bucharest': ['Fagaras', 'Pitesti']
}

# Distances
cost = {
    ('Arad','Sibiu'):140, ('Sibiu','Fagaras'):99, ('Fagaras','Bucharest'):211
}

def bfs(start, goal):
    queue = deque([[start]])
    visited = set()

    while queue:
        path = queue.popleft()
        node = path[-1]

        if node == goal:
            return path

        if node not in visited:
            visited.add(node)
            for neighbor in graph[node]:
                new_path = list(path)
                new_path.append(neighbor)
                queue.append(new_path)

    return None

# Run BFS
path = bfs('Arad', 'Bucharest')

# Calculate distance
total_distance = 0
for i in range(len(path)-1):
    total_distance += cost.get((path[i], path[i+1]), 0)

print("Path:", " -> ".join(path))
print("Total Distance:", total_distance)
```
